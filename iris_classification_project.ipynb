**TRANING AND TESTING THE DATA/MODEL**

```python
x_train=iris.drop("class",axis=1)
y_train=iris["class"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
act=accuracy_score(y_train,log.predict(x_train))
print("train accuracy is:",(act*100))
act=accuracy_score(y_test,log.predict(x_test))
print("test accuracy is:",(act*100))
```

**the train accuracy is 96.39**
**the test accuracy is 100.0**
